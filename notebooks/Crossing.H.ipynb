# Quantile crossing validation of GPQR models for H

This notebook visualizes quantile crossing from direct GPQR.

GPR and center-gap GPQR are not tested because they are mathematically immune to quantile crossing.

The result shows that direct GPQRs are susceptible to quantile crossing.
Although crossings disappear in higher epochs, the number of required epochs varies by input prediction points.

Hence, direct GPQRs should be avoided for general purpose models.

```python
import sys
import os

import pandas as pd
import matplotlib.pyplot as plt

sys.path.insert(0, os.path.abspath(".."))

try:
    import config_notebook
except ImportError:
    print("Output will not be deterministic SVG.")
```

## Load data

```python
direct_lmc_qc = pd.read_csv("../_temp/crossing.DirectLmcMtgpqr_H.csv")
direct_independent_qc = pd.read_csv("../_temp/crossing.DirectIndependentMtgpqr_H.csv")
```

## Plot

```python
plt.plot(direct_lmc_qc["crossing_rate_0"], color="tab:blue", label="Direct LMC")
plt.plot(direct_lmc_qc["crossing_rate_1"], color="tab:blue", alpha=0.5)
plt.plot(
    direct_independent_qc["crossing_rate_0"],
    color="tab:orange",
    label="Direct Independent",
)
plt.plot(direct_independent_qc["crossing_rate_1"], color="tab:orange", alpha=0.5)
plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Crossing Rate")
plt.legend()
plt.show()
```

```python
plt.plot(direct_lmc_qc["mean_crossing_0"], color="tab:blue", label="Direct LMC")
plt.plot(direct_lmc_qc["mean_crossing_1"], color="tab:blue", alpha=0.5)
plt.plot(
    direct_independent_qc["mean_crossing_0"],
    color="tab:orange",
    label="Direct Independent",
)
plt.plot(direct_independent_qc["mean_crossing_1"], color="tab:orange", alpha=0.5)
plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Mean crossing")
plt.legend()
plt.show()
```

```python
plt.plot(direct_lmc_qc["max_crossing_0"], color="tab:blue", label="Direct LMC")
plt.plot(direct_lmc_qc["max_crossing_1"], color="tab:blue", alpha=0.5)
plt.plot(
    direct_independent_qc["max_crossing_0"],
    color="tab:orange",
    label="Direct Independent",
)
plt.plot(direct_independent_qc["max_crossing_1"], color="tab:orange", alpha=0.5)
plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Max crossing")
plt.legend()
plt.show()
```
